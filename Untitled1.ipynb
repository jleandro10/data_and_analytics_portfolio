## Detecção de Fraudes em Transações com Cartão de Crédito

<p>Fraudes em transações com cartão de crédito são motivo de preocupações principalmente para instituições financeiras. Com a pandemia, a população passou a fazer mais compras pela internet, fato que contribuiu para o aumento de fraudes.</p> 
<p>De acordo com reportagem veiculada no Infomoney no dia 21 de julho de 2021, disponível <a href="https://www.infomoney.com.br/minhas-financas/brasil-teve-alta-de-quase-33-nas-tentativas-de-fraude-com-cartao-de-credito-no-1-semestre-mostra-estudo/">aqui</a>, a empresa Clearsale demonstrou em um estudo que as tentativas de fraude com cartão de crédito em e-commerce aumentaram 32,7% no primeiro semestre de 2021 em comparação com o mesmo período do ano de 2020.</p>
<p>Diante das informações acima, um bom modelo de machine learning pode fazer a diferença. Detectar fraudes usando uma base de dados de transações com cartão de crédito pode representar economia de muito dinheiro</p>
<p>O presente projeto visa criar modelos de machine learning capazes de fazer a detecção de fraudes a patir de um banco de dados de transações com cartão de crédito e compará-los.</p>

## Tratamento dos Dados

<p>Antes de criar um modelo de classificação, vamos padronizar Time e Amount, fazer a divisão dos dados entre treino e teste e balancear os dados.</p>

Abaixo temos a padronização de Time e Amount

```python
from sklearn.preprocessing import StandardScaler

df_std = df.copy()

scaler = StandardScaler()
df_std['std_amount'] = scaler.fit_transform(df_std['Amount'].values.reshape(-1,1))
df_std['std_time'] = scaler.fit_transform(df_std['Time'].values.reshape(-1,1))

df_std.drop(['Time', 'Amount'], axis=1, inplace=True)

df_std.head()
```

## Divisão dos dados

<p>Abaixo temos a divisão dos dados em treino e teste</p>
<p>Uma observação a ser feita é que foi usado o parâmetro stratify=y a fim de dividir os dados de treino e teste respeitando a proporção das classificações disponíveis no dataset, buscando garantir uma melhor performance para o modelo.</p>

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, roc_auc_score, roc_curve, accuracy_score

X = df.drop('Class', axis=1)
y = df['Class']

X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, shuffle=True)
```

## Balanceamento dos dados

<p>O balanceamento foi feito usando under sampling, o que permite preservar a classe que possui menos dados</p>
<p>Inspecionando o gráfico, percebe-se que não há mais discrepância entre as quantidades para cada classe.</p>

```python
# Balancear o conjunto de dados
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler()
X_rus, y_rus = rus.fit_resample(X_train, y_train)

sns.countplot(y_rus)
```

## Modelos de Machine Learning

<p>Serão criados 3 modelos de machine learning a fim de classificar as transações em fraudulentas e não fraudulentas: regressão logística, árvore de decisão e xgboost.</p>
<p>Os modelos comparados no final do processo.</p>

### Regressão Logística

Segue abaixo o modelo de regressão logística

```python
from sklearn.linear_model import LogisticRegression

lg_model = LogisticRegression()
lg_model.fit(X_rus, y_rus)

y_pred = lg_model.predict(X_test)
```

<p>Após a definição do modelo, treino e previsões, segue abaixo a avaliação do desempenho.</p>
<p>O modelo de regressão logística teve um bom desempenho, atingindo acurácia de 96%.</p>
<p>A matriz de confusão também mostra o bom desempenho do modelo, obtendo uma boa taxa de acertos.</p>
<p>Outra métrica utilizada é a área sobre a curva (AUC). O valor obtido nesta métrica também corrobora com o bom desempenho.</p> 

```python
import scikitplot as skplt

# matriz de confusão
skplt.metrics.plot_confusion_matrix(y_test, y_pred, normalize=True)

# relatório da classificação
print("Relatório de Classificação: \n", classification_report(y_test, y_pred))
```

```python
# acurácia
print("Acurácia: {:.4f}".format(accuracy_score(y_test, y_pred)))

# área sobre a curva
print("AUC: {:.4f}".format(roc_auc_score(y_test, y_pred)))
```

### Árvore de Decisão

Segue abaixo o modelo da árvore de decisão

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(max_depth=3, criterion="entropy")
dt_model.fit(X_rus, y_rus)

y_pred_dt = dt_model.predict(X_test)
```

<p>O modelo de árvore de decisão também obteve um bom desempenho, embora ligeiramente inferior ao modelo de regressão logística. Isso pode ser explicado pela parametrização do modelo, sendo necessário maior aprofundamento neste tema.</p>
<p>As mesmas métricas aplicadas ao modelo anterior foram aplicadas aqui:</p>


*   Acurácia de 95%;
*   Área sobre a curva de 93%;
*   A matriz de confusão também mostra uma boa taxa de acertos.





```python
# matriz de confusão
skplt.metrics.plot_confusion_matrix(y_test, y_pred_dt, normalize=True)

# relatório de classificação
print("Relatório de Classificação: \n", classification_report(y_test, y_pred_dt))
```

```python
# acurácia
print("Acurácia: {:.4f}".format(accuracy_score(y_test, y_pred_dt)))

# área sobre a curva
print("AUC: {:.4f}".format(roc_auc_score(y_test, y_pred_dt)))
```
